### Objective

```python
from cs103 import *
from typing import NamedTuple, List, Optional
from enum import Enum
import csv
import matplotlib.pyplot as pyplot

##################
# Data Definitions

StateResults = NamedTuple('StateResults', [('state', str), # 56 different cases
                                           ('biden_poll_day', float), # in range [0.00, 100.00] as percent
                                           ('trump_poll_day', float), # in range [0.00, 100.00] as percent
                                           ('biden_poll_week', float), # in range [0.00, 100.00] as percent
                                           ('trump_poll_week', float), # in range [0.00, 100.00] as percent                                                                                     
                                           ('biden_poll_month', float), # in range [0.00, 100.00] as percent
                                           ('trump_poll_month', float), # in range [0.00, 100.00] as percent                                         
                                           ('biden_actual', float), # in range [0.00, 100.00] as percent
                                           ('trump_actual', float)]) # in range [0.00, 100.00] as percent

# interpret the actual vote percentage and the polling estimate percentage (at 1 day before, 1 week before,
# 1 month before the election) that a candidate receives in a state

SRWY = StateResults('Wyoming', 30.81, 61.97, 28.80, 65.69, 28.30, 67.76, 26.60, 69.90)
SRWI = StateResults('Wisconsin', 52.13, 43.67, 51.38, 44.14, 50.51, 43.91, 49.45, 48.83)
SRWV = StateResults('West Virginia', 33.49, 61.90, 35.65, 59.41, 32.51, 63.75, 29.70, 68.60)
SRWA = StateResults('Washington', 59.34, 36.26, 58.21, 35.13, 59.76, 34.05, 58.40, 39.00)

SRCT = StateResults('Connecticut', 58.39, 31.96, 57.90, 33.64, 56.37, 34.63, 59.20, 39.20)
SRCO = StateResults('Colorado', 53.63, 41.00, 53.58, 39.68, 51.95, 40.42, 55.40, 41.90)
SRCA = StateResults('California', 61.63, 32.29, 60.97, 31.52, 61.43, 31.23, 63.50, 34.30)

@typecheck
# template based on Compound
def fn_for_state_results(sr: StateResults) -> ...:
    return ...(sr.state,
               sr.biden_poll_day,
               sr.trump_poll_day,
               sr.biden_poll_week,
               sr.trump_poll_week,
               sr.biden_poll_month,
               sr.trump_poll_month,
               sr.biden_actual,
               sr.trump_actual)

# List[StateResults]
# interpret a list of state results

LOSR0 = []
LOSR1 = [SRWY, SRWI, SRWV, SRWA]
LOSR2 = [SRCT, SRCO, SRCA]

# template based on arbitrary-sized and the reference rule 
@typecheck
def fn_for_losr(losr: List[StateResults]) -> ...:
    # description of the acc
    acc = ... # type: ...
    for sr in losr:
        acc = ...(fn_for_state_results(sr), acc)
    return ...(acc)

StateMargins = NamedTuple('StateMargins', [('state', str), # 56 different cases
                                           ('poll_day_margin', float), # in range [-100.00, 100.00] 
                                                                       # as a percentage
                                           ('poll_week_margin', float), # in range [-100.00, 100.00] 
                                                                        # as a percentage
                                           ('poll_month_margin', float), # in range [-100.00, 100.00] 
                                                                         # as a percentage
                                           ('actual_margin', float)]) # in range [-100.00, 100.00] 
                                                                      # as a percentage

# interpret the difference between the actual vote percentage for Biden and for Trump in a state 
# and the difference between the polling estimate percentage (at 1 day before, 1 week before,
# 1 month before the election) for Biden and for Trump in a state.
# Negative numbers mean Biden wins a state, positive numbers mean Trump wins a state

SMWY = StateMargins('Wyoming', 31.16, 36.89, 39.46, 43.30)
SMWI = StateMargins('Wisconsin', -8.46, -7.24, -6.60, -0.62)
SMWV = StateMargins('West Virginia', 28.41, 23.76, 31.24, 38.90)
SMWA = StateMargins('Washington', -23.08, -23.08, -25.71, -19.40)

SMCT = StateMargins('Connecticut', -26.43, -24.26, -21.74, -20.00)
SMCO = StateMargins('Colorado', -12.63, -13.90, -11.53, -13.50)
SMCA = StateMargins('California', -29.34, -29.45, -30.20, -29.20)

@typecheck
# template based on Compound
def fn_for_state_margins(sm: StateMargins) -> ...:
    return ...(sm.state,
               sm.poll_day_margin,
               sm.poll_week_margin,
               sm.poll_month_margin,
               sm.actual_margin)

# List[StateMargins]
# interpret a list of state margins

LOSM0 = []
LOSM1 = [SMWY, SMWI, SMWV, SMWA]
LOSM2 = [SMCT, SMCO, SMCA]

# template based on arbitrary-sized and the reference rule 
@typecheck
def fn_for_losm(losm: List[StateMargins]) -> ...:
    # description of the acc
    acc = ... # type: ...
    for sm in losm:
        acc = ...(fn_for_state_margins(sm), acc)
    return ...(acc)

# List[float] 
# interpret a list of numbers
LOF0 = []
LOF1 = [1.0, 2.0]
LOF2 = [-1.1, -2.2, 3.3, 4.4]

# template based on arbitrary-sized
@typecheck
def fn_for_lof(lof: List[float]) -> ...:
    # description of the acc
    acc = ... # type:
    for f in lof:
        ...(f, acc)
    return ...(acc)

# List[str] 
# interpret a list of strings
LOS0 = []
LOS1 = ["Hello", "hi"]
LOS2 = ["state", "result", "margin"]

# template based on arbitrary-sized
@typecheck
def fn_for_los(los: List[str]) -> ...:
    # description of the acc
    acc = ... # type:
    for s in los:
        ...(s, acc)
    return ...(acc)
```

#### Read and analyze functions to handle information and analyze the data

```python
###########
# Functions

@typecheck
def main(filename: str) -> None:
    """
    Reads the file from given filename, plots the outcome from the margin difference function
    and returns None
    """
#   Template from HtDAP, based on function composition 
    return show_scatterplot(margin_differences(read(filename))) 

@typecheck
def show_scatterplot(losm: List[StateMargins]) -> None:
    """
    display a scatterplot of actual result margins annd polling result margins (at 1 day before, 
    1 week before, 1 month before the election) in losm
    """
#   return None # stub
#   template from visualization
    pyplot.figure(figsize = (10, 5))
    pyplot.xlabel("Actual Margin")
    pyplot.ylabel("Polling Margin")
    pyplot.title("Polling Performance Against Election Results")
    
    pyplot.scatter(list_poll_day_margins(losm), list_actual_margins(losm), marker = 'o', c = 'b', 
                   label = str(average_day_divergance(losm)) + 
                   " average deviation from polls day before election")
    
    pyplot.scatter(list_poll_week_margins(losm), list_actual_margins(losm), marker = 'o', c = 'r', 
                   label = str(average_week_divergance(losm)) + 
                   " average deviation from polls week before election")
    
    pyplot.scatter(list_poll_month_margins(losm), list_actual_margins(losm), marker = 'o', c = 'g', 
                   label = str(average_month_divergance(losm)) + 
                   " average deviation from polls month before election")

    pyplot.legend(loc = 'best')
    
    pyplot.grid()
    
    pyplot.show()
    
    return None

@typecheck
def average_day_divergance(losm: List[StateMargins]) -> float:
    """
    calculates the average difference between all the state polls (poll average from day before election) 
    and the actual results
    """
#   return 0.0 # stub
#   template from List[StateMargins]
    # total_day_divergence stores the sum of the divergences seen so far
    total_day_divergence = 0 # type: float
    for sm in losm:
        total_day_divergence = total_day_divergence + get_day_divergence(sm)
    
    if len(losm) == 0:
        return 0
    else:
        return round(total_day_divergence/len(losm), 2)
    
@typecheck
def get_day_divergence(sm: StateMargins) -> float:
    """
    return the divergence (poll average from day before election) from a given state margin
    """
#   return 0.0 # stub
    # template from StateMargin
    return abs(sm.poll_day_margin - sm.actual_margin)

@typecheck
def average_week_divergance(losm: List[StateMargins]) -> float:
    """
    calculates the average difference between all the state polls (poll average from week before election) 
    and the actual results
    """
#   return 0.0 # stub
#   template from List[StateMargins]
    # total_week_divergence stores the sum of the divergences seen so far
    total_week_divergence = 0 # type: float
    for sm in losm:
        total_week_divergence = total_week_divergence + get_week_divergence(sm)
    
    if len(losm) == 0:
        return 0
    else:
        return round(total_week_divergence/len(losm), 2)
    
@typecheck
def get_week_divergence(sm: StateMargins) -> float:
    """
    return the divergence (poll average from week before election) from a given state margin
    """
#   return 0.0 # stub
    # template from StateMargin
    return abs(sm.poll_week_margin - sm.actual_margin)

@typecheck
def average_month_divergance(losm: List[StateMargins]) -> float:
    """
    calculates the average difference between all the state polls (poll average from month before election) 
    and the actual results
    """
#   return 0.0 # stub
#   template from List[StateMargins]
    # total_month_divergence stores the sum of the divergences seen so far
    total_month_divergence = 0 # type: float
    for sm in losm:
        total_month_divergence = total_month_divergence + get_month_divergence(sm)
    
    if len(losm) == 0:
        return 0
    else:
        return round(total_month_divergence/len(losm), 2)
    
@typecheck
def get_month_divergence(sm: StateMargins) -> float:
    """
    return the divergence (poll average from month before election) from a given state margin
    """
#   return 0.0 # stub
    # template from StateMargin
    return abs(sm.poll_month_margin - sm.actual_margin)

@typecheck
def margin_differences(losr: List[StateResults]) -> List[StateMargins]:
    """ 
    return a list of state margins from a list of state results as a percentage.
    A margin is the difference between the winning candidate and a loosing candidate.
    """ 
#   return [] # stub
#   template based on List[StateResults]
    # acc contains all the state margins seen so far
    acc = [] # type: List[StateMargins]
    for sr in losr:
        acc.append(get_margins(sr))
    return acc

@typecheck
def get_margins(sr: StateResults) -> StateMargins:
    """
    return the state margins for a given state results
    """
#   return SMWA # stub
#   template from StateResults
    sm = StateMargins(sr.state, get_poll_day_margin(sr), get_poll_week_margin(sr),
                      get_poll_month_margin(sr), get_actual_margin(sr))
    return sm

@typecheck
def get_poll_day_margin(sr: StateResults) -> float:
    """
    return the polling result margin (poll average from day before election) for a given state results
    """
#   return 0.0 # stub
#   template from StateResults
    return round((sr.trump_poll_day - sr.biden_poll_day), 2)

@typecheck
def get_poll_week_margin(sr: StateResults) -> float:
    """
    return the polling result margin (poll average from week before election) for a given state results
    """
#   return 0.0 # stub
#   template from StateResults
    return round((sr.trump_poll_week - sr.biden_poll_week), 2)

@typecheck
def get_poll_month_margin(sr: StateResults) -> float:
    """
    return the polling result margin (poll average from month before election) for a given state results
    """
#   return 0.0 # stub
#   template from StateResults
    return round((sr.trump_poll_month - sr.biden_poll_month), 2)
       
@typecheck
def get_actual_margin(sr: StateResults) -> float:
    """
    return the polling result margin for a given state result
    """
#   return 0.0 # stub
#   template from StateResults
    return round((sr.trump_actual - sr.biden_actual), 2)

@typecheck
def list_poll_day_margins(losm: List[StateMargins]) -> List[float]:
    """
    return a list of poll margins (polling from day before election) from losm
    """
#   return [] # stub
#   template from List[StateMargins]
    # poll_list_day contains all the heights seen so far
    poll_list_day = [] # type: List[float]
    for sm in losm:
        poll_list_day.append(sm.poll_day_margin) 
    return poll_list_day

@typecheck
def list_poll_week_margins(losm: List[StateMargins]) -> List[float]:
    """
    return a list of poll margins (polling from week before election) from losm
    """
#   return [] # stub
#   template from List[StateMargins]
    # poll_list_week contains all the heights seen so far
    poll_list_week = [] # type: List[float]
    for sm in losm:
        poll_list_week.append(sm.poll_week_margin) 
    return poll_list_week

@typecheck
def list_poll_month_margins(losm: List[StateMargins]) -> List[float]:
    """
    return a list of poll margins (polling from month before election) from losm
    """
#   return [] #stub
#   template from List[StateMargins]
    # poll_list_month contains all the heights seen so far
    poll_list_month = [] # type: List[float]
    for sm in losm:
        poll_list_month.append(sm.poll_month_margin) 
    return poll_list_month
                
@typecheck
def list_actual_margins(losm: List[StateMargins]) -> List[float]:
    """
    return a list of actual margins from losm
    """
#   return [] # stub
#   template from List[StateMargins]
    # weight_list contains all the heights seen so far
    actual_list = [] # type: List[float]
    for sm in losm:
        actual_list.append(sm.actual_margin)
    return actual_list

@typecheck
def list_states(losm: List[StateMargins]) -> List[str]:
    """
    return a list of states from losm
    """
#   return [] # stub
#   template from List[StateMargins]
    # weight_list contains all the heights seen so far
    state_list = [] # type: List[float]
    for sm in losm:
        state_list.append(sm.state)
    return state_list

@typecheck
def read(filename: str) -> List[StateResults]:
    """    
    reads information from the specified file and returns a list of state results
    """
#   return []  #stub
#   Template from HtDAP
#   losr contains the result so far
    losr = [] # type: List[StateResults]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            sr = StateResults(row[0], round(parse_float(row[3]), 2), round(parse_float(row[4]), 2),
                              round(parse_float(row[5]), 2), round(parse_float(row[6]), 2),
                              round(parse_float(row[13]), 2), round(parse_float(row[14]), 2),
                              round(parse_float(row[1]), 2), round(parse_float(row[2]), 2))
            losr.append(sr)
    
    return losr

start_testing()

# Examples and tests for main
expect(main('presidential_poll_averages_test1.csv'), None)
expect(main('presidential_poll_averages_test2.csv'), None)

# Examples and tests for show_scatterplot
expect(show_scatterplot(LOSM0), None)
expect(show_scatterplot(LOSM1), None)
expect(show_scatterplot(LOSM2), None)

# Examples and tests for average_day_divergance
expect(average_day_divergance(LOSM0), 0)
expect(average_day_divergance(LOSM1), 8.54)
expect(average_day_divergance(LOSM2), 2.48)

# Examples and tests for get_day_divergence
expect(get_day_divergence(SMCO), 0.87)
expect(get_day_divergence(SMWV), 10.49)

# Examples and tests for average_week_divergance
expect(average_week_divergance(LOSM0), 0)
expect(average_week_divergance(LOSM1), 7.96)
expect(average_week_divergance(LOSM2), 1.64)

# Examples and tests for get_week_divergence
expect(get_week_divergence(SMCO), 0.40)
expect(get_week_divergence(SMWV), 15.14)

# Examples and tests for average_month_divergance
expect(average_month_divergance(LOSM0), 0)
expect(average_month_divergance(LOSM1), 5.95)
expect(average_month_divergance(LOSM2), 1.57)

# Examples and tests for get_month_divergence
expect(get_month_divergence(SMCO), 1.97)
expect(get_month_divergence(SMWV), 7.66)

# Examples and tests for margin_differences 
expect(margin_differences(LOSR0), LOSM0)
expect(margin_differences(LOSR1), LOSM1)
expect(margin_differences(LOSR2), LOSM2)

# Examples and tests for get_margins
expect(get_margins(SRCO), SMCO)
expect(get_margins(SRWV), SMWV)

# Examples and tests for get_poll_day_margin
expect(get_poll_day_margin(SRCO), -12.63)
expect(get_poll_day_margin(SRWV), 28.41)

# Examples and tests for get_poll_week_margin
expect(get_poll_week_margin(SRCO), -13.90)
expect(get_poll_week_margin(SRWV), 23.76)

# Examples and tests for get_poll_month_margin
expect(get_poll_month_margin(SRCO), -11.53)
expect(get_poll_month_margin(SRWV), 31.24)

# Examples and tests for get_actual_margin
expect(get_actual_margin(SRCO), -13.5)
expect(get_actual_margin(SRWV), 38.90)

# Examples and tests for list_poll_day_margins
expect(list_poll_day_margins(LOSM0), [])
expect(list_poll_day_margins(LOSM1), [31.16, -8.46, 28.41, -23.08])
expect(list_poll_day_margins(LOSM2), [-26.43, -12.63, -29.34])

# Examples and tests for list_poll_week_margins
expect(list_poll_week_margins(LOSM0), [])
expect(list_poll_week_margins(LOSM1), [36.89, -7.24, 23.76, -23.08])
expect(list_poll_week_margins(LOSM2), [-24.26, -13.90, -29.45])

# Examples and tests for list_poll_month_margins
expect(list_poll_month_margins(LOSM0), [])
expect(list_poll_month_margins(LOSM1), [39.46, -6.60, 31.24, -25.71])
expect(list_poll_month_margins(LOSM2), [-21.74, -11.53, -30.20])

# Examples and tests for list_actual_margins
expect(list_actual_margins(LOSM0), [])
expect(list_actual_margins(LOSM1), [43.30, -0.62, 38.90, -19.40])
expect(list_actual_margins(LOSM2), [-20, -13.50, -29.20])
                
# Examples and tests for list_states
expect(list_states(LOSM0), [])
expect(list_states(LOSM1), ['Wyoming', 'Wisconsin', 'West Virginia', 'Washington'])
expect(list_states(LOSM2), ['Connecticut', 'Colorado', 'California'])
                
# Examples and tests for read
expect(read('presidential_poll_averages_test1.csv'), [SRWY, SRWI, SRWV, SRWA])
expect(read('presidential_poll_averages_test2.csv'), [SRCT, SRCO, SRCA])

summary()
```
